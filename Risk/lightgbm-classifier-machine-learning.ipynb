# Analysis of Categorical and Numerical Variables

```python
def grab_col_names(dataframe, cat_th=10, car_th=20):
  #Catgeorical Variable Selection
  cat_cols = [col for col in dataframe.columns if str(dataframe[col].dtypes) in ["category","object","bool"]]
  num_but_cat = [col for col in dataframe.columns if dataframe[col].nunique() < cat_th and dataframe[col].dtypes in ["uint8","int64","float64"]]
  cat_but_car = [col for col in dataframe.columns if dataframe[col].nunique() > car_th and str(dataframe[col].dtypes) in ["category","object"]]
  cat_cols = cat_cols + num_but_cat
  cat_cols = [col for col in cat_cols if col not in cat_but_car]

  #Numerical Variable Selection
  num_cols = [col for col in dataframe.columns if dataframe[col].dtypes in ["uint8","int64","float64"]]
  num_cols = [col for col in num_cols if col not in cat_cols]

  return cat_cols, num_cols, cat_but_car, num_but_cat
```

```python
def high_correlated_cols(dataframe, corr_th = 0.90, plot=False):
  num_cols = [col for col in dataframe.columns if dataframe[col].dtypes in ["uint8", "int64", "float64"]]
  corr = dataframe[num_cols].corr()
  corr_matrix = corr.abs()
  upper_triangular_matrix = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
  drop_list = [col for col in upper_triangular_matrix.columns if any(upper_triangular_matrix[col] > corr_th)]
  if drop_list == []:
    print("Aftre corelation analysis, we dont need to remove variables")

  if plot:
    sns.set(rc={'figure.figsize': (14,7)})
    sns.heatmap(corr, cmap='RdBu', annot=True, fmt=".2f")
    plt.show()

  return drop_list
```

# Outlier Analysis

```python
def outlier_thresholds(dataframe,col_name,q1=0.05,q3=0.95):
  quartile1 = dataframe[col_name].quantile(q1)
  quartile3 = dataframe[col_name].quantile(q3)
  interquartile_range = quartile3 - quartile1
  low_limit = quartile1 - 1.5 * interquartile_range
  up_limit = quartile3 + 1.5 * interquartile_range
  return low_limit,up_limit
```

```python
def check_outlier(dataframe, col_name):
  low_limit,up_limit = outlier_thresholds(dataframe,col_name)
  if dataframe[(dataframe[col_name] > up_limit) | (dataframe[col_name] < low_limit)].any(axis=None):
    return True
  else:
    return False
```

```python
def replace_with_thresholds(dataframe, col_name):
  low_limit, up_limit = outlier_thresholds(dataframe, col_name)
  dataframe.loc[(dataframe[col_name] < low_limit), col_name] = low_limit
  dataframe.loc[(dataframe[col_name] > up_limit), col_name] = up_limit
```

```python
for col in df.columns:
  print(col, check_outlier(df, col))
  if check_outlier(df, col):
    replace_with_thresholds(df, col)
```

# Create a Base Model LightGBM Classifier

**LightGBM**: LightGBM (Light Gradient Boosting Machine) is a variant of the Gradient Boosting framework and is an open source machine learning library that provides fast and high-performance modeling on large data sets. LightGBM is an algorithm developed by Microsoft and gaining popularity.
In the LightGBM algorithm, the "leaf-wise" tree growth strategy represents a different approach from the traditional "level-wise" tree growth strategy. This strategy used in LightGBM focuses on making the best split at each step while growing the tree, resulting in deeper and more complex tree structures.

```python
def LGBM_Model(dataframe, target, test_size=0.30, cv=10, results=False, conf_matrix=False, c_report=False, roc=False, plot_importance=False, save_model=False):
  X = dataframe.drop(target, axis=1)
  y = dataframe[target]
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=1)

  lgbm_model = LGBMClassifier(verbose=-1).fit(X_train, y_train)

  if results:
    acc_train = accuracy_score(y_train, lgbm_model.predict(X_train))
    acc_test = accuracy_score(y_test, lgbm_model.predict(X_test))
    r2 = lgbm_model.score(X, y)
    r2_train = lgbm_model.score(X_train, y_train)
    r2_test = lgbm_model.score(X_test, y_test)
    cv_train = cross_val_score(lgbm_model, X_train, y_train, cv=cv).mean()
    cv_test = cross_val_score(lgbm_model, X_test, y_test, cv=cv).mean()
    cv_results = cross_validate(lgbm_model, X, y, cv=cv, scoring=["accuracy", "precision", "recall", "f1", "roc_auc"])
    print("##################### Accuracy & R2 Metrics ######################")
    print("Accuracy_train: ", "%.3f" % acc_train)
    print("Accuracy_test: ", "%.3f" % acc_test)
    print("R2: ", "%.3f" % r2)
    print("R2_train: ", "%.3f" % r2_train)
    print("R2_test: ", "%.3f" % r2_test)
    print("Cross Validation Score(Train): ", "%.3f" % cv_train)
    print("Cross Validation Score(Test): ", "%.3f" % cv_test)
    print("Cross Validate (Accuracy): ", "%.3f" % cv_results['test_accuracy'].mean())
    print("Cross Validate (Precision): ", "%.3f" % cv_results['test_precision'].mean())
    print("Cross Validate (Recall): ", "%.3f" % cv_results['test_recall'].mean())
    print("Cross Validate (F1): ", "%.3f" % cv_results['test_f1'].mean())
    print("Cross Validate (ROC AUC Score): ", "%.3f" % cv_results['test_roc_auc'].mean())

  if conf_matrix:
    print("##################### Confusion Matrix(Train) ######################")
    print(confusion_matrix(y_train, lgbm_model.predict(X_train)))
    print("##################### Confusion Matrix(Test) ######################")
    print(confusion_matrix(y_test, lgbm_model.predict(X_test)))
    print("##################### Confusion Matrix(All) ######################")
    print(confusion_matrix(y, lgbm_model.predict(X)))
    plt.figure(figsize=(5,5))
    acc = accuracy_score(y, lgbm_model.predict(X))
    cm = confusion_matrix(y, lgbm_model.predict(X))
    sns.heatmap(cm, annot=True, fmt=".0f")
    plt.xlabel('y_pred')
    plt.ylabel('y')
    plt.title('Accuracy Score: {}'.format(acc), size=10)
    plt.show()

  if c_report:
    print("###################### Classification Report(Train) ######################")
    print(classification_report(y_train, lgbm_model.predict(X_train)))
    print("###################### Classification Report(Test) ######################")
    print(classification_report(y_test, lgbm_model.predict(X_test)))
    print("###################### Classification Report(All) ######################")
    print(classification_report(y, lgbm_model.predict(X)))

  if roc:
    print("###################### Roc Auc Score(Train) ######################")
    lgbm_roc_auc_train = roc_auc_score(y_train, lgbm_model.predict(X_train))
    print("Roc Auc Train: ", "%.3f" % lgbm_roc_auc_train)
    fpr, tpr, thresholds = roc_curve(y_train, lgbm_model.predict_proba(X_train)[:,1])
    plt.figure(figsize=(5,5))
    plt.plot(fpr, tpr, label='AUC (area = %.3f)' % lgbm_roc_auc_train )
    plt.plot([0, 1], [0, 1], 'r--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Ratio')
    plt.ylabel('True Positive Ratio')
    plt.title('ROC')
    plt.show()

    print("###################### Roc Auc Score(Test) ######################")
    lgbm_roc_auc_test = roc_auc_score(y_test, lgbm_model.predict(X_test))
    print("Roc Auc Test: ", "%.3f" % lgbm_roc_auc_test)
    fpr, tpr, thresholds = roc_curve(y_test, lgbm_model.predict_proba(X_test)[:,1])
    plt.figure(figsize=(5,5))
    plt.plot(fpr, tpr, label='AUC (area = %.3f)' % lgbm_roc_auc_test )
    plt.plot([0, 1], [0, 1], 'r--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Ratio')
    plt.ylabel('True Positive Ratio')
    plt.title('ROC')
    plt.show()

  if plot_importance:
    feature_imp = pd.DataFrame({'Value': lgbm_model.feature_importances_, 'Feature': X.columns})
    plt.figure(figsize=(8, 8))
    sns.set(font_scale=1)
    sns.barplot(x="Value", y="Feature", data=feature_imp.sort_values(by="Value",
                                                                     ascending=False)[0:len(X)])
    plt.title('Features')
    plt.tight_layout()
    plt.savefig('importances.png')
    plt.show()

  if save_model:
    joblib.dump(lgbm_model, "lgbm_model.pkl")

  return lgbm_model
```

# Create a Model Tuning: LGBM Classifier

```python
def LGBM_Model_Tuned(dataframe, target, test_size=0.30, cv=10, results=False, conf_matrix=False, c_report=False, roc=False, plot_importance=False, save_model=False):
  X = dataframe.drop(target, axis=1)
  y = dataframe[target]
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=1)

  lgbm_model = LGBMClassifier(verbose=-1)
  lgbm_params = {
    "n_estimators": [500, 1000],
    "subsample": [0.6, 0.8],
    "max_depth": [5, 8],
    "learning_rate": [0.1, 0.01],
    "min_child_samles": [5, 10]
  }
  lgbm_cv_model = GridSearchCV(lgbm_model, lgbm_params, cv=cv, verbose=True, n_jobs=-1).fit(X_train, y_train)
  lgbm_model_tuned = lgbm_model.set_params(**lgbm_cv_model.best_params_).fit(X, y)

  if results:
    acc_train = accuracy_score(y_train, lgbm_model_tuned.predict(X_train))
    acc_test = accuracy_score(y_test, lgbm_model_tuned.predict(X_test))
    r2 = lgbm_model_tuned.score(X, y)
    r2_train = lgbm_model_tuned.score(X_train, y_train)
    r2_test = lgbm_model_tuned.score(X_test, y_test)
    cv_train = cross_val_score(lgbm_model_tuned, X_train, y_train, cv=cv).mean()
    cv_test = cross_val_score(lgbm_model_tuned, X_test, y_test, cv=cv).mean()
    cv_results = cross_validate(lgbm_model_tuned, X, y, cv=cv, scoring=["accuracy", "precision", "recall", "f1", "roc_auc"])
    print("##################### Accuracy & R2 Metrics ######################")
    print("LGBM Classifier Best Params: ", lgbm_cv_model.best_params_)
    print("Accuracy_train: ", "%.3f" % acc_train)
    print("Accuracy_test: ", "%.3f" % acc_test)
    print("R2: ", "%.3f" % r2)
    print("R2_train: ", "%.3f" % r2_train)
    print("R2_test: ", "%.3f" % r2_test)
    print("Cross Validation Score(Train): ", "%.3f" % cv_train)
    print("Cross Validation Score(Test): ", "%.3f" % cv_test)
    print("Cross Validate (Accuracy): ", "%.3f" % cv_results['test_accuracy'].mean())
    print("Cross Validate (Precision): ", "%.3f" % cv_results['test_precision'].mean())
    print("Cross Validate (Recall): ", "%.3f" % cv_results['test_recall'].mean())
    print("Cross Validate (F1): ", "%.3f" % cv_results['test_f1'].mean())
    print("Cross Validate (ROC AUC Score): ", "%.3f" % cv_results['test_roc_auc'].mean())

  if conf_matrix:
    print("##################### Confusion Matrix(Train) ######################")
    print(confusion_matrix(y_train, lgbm_model_tuned.predict(X_train)))
    print("##################### Confusion Matrix(Test) ######################")
    print(confusion_matrix(y_test, lgbm_model_tuned.predict(X_test)))
    print("##################### Confusion Matrix(All) ######################")
    print(confusion_matrix(y, lgbm_model_tuned.predict(X)))
    plt.figure(figsize=(5,5))
    acc = accuracy_score(y, lgbm_model_tuned.predict(X))
    cm = confusion_matrix(y, lgbm_model_tuned.predict(X))
    sns.heatmap(cm, annot=True, fmt=".0f")
    plt.xlabel('y_pred')
    plt.ylabel('y')
    plt.title('Accuracy Score: {}'.format(acc), size=10)
    plt.show()

  if c_report:
    print("###################### Classification Report(Train) ######################")
    print(classification_report(y_train, lgbm_model_tuned.predict(X_train)))
    print("###################### Classification Report(Test) ######################")
    print(classification_report(y_test, lgbm_model_tuned.predict(X_test)))
    print("###################### Classification Report(All) ######################")
    print(classification_report(y, lgbm_model_tuned.predict(X)))

  if roc:
    print("###################### Roc Auc Score(Train) ######################")
    lgbm_roc_auc_train = roc_auc_score(y_train, lgbm_model_tuned.predict(X_train))
    print("Roc Auc Train: ", "%.3f" % lgbm_roc_auc_train)
    fpr, tpr, thresholds = roc_curve(y_train, lgbm_model_tuned.predict_proba(X_train)[:,1])
    plt.figure(figsize=(5,5))
    plt.plot(fpr, tpr, label='AUC (area = %.3f)' % lgbm_roc_auc_train )
    plt.plot([0, 1], [0, 1], 'r--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Ratio')
    plt.ylabel('True Positive Ratio')
    plt.title('ROC')
    plt.show()

    print("###################### Roc Auc Score(Test) ######################")
    lgbm_roc_auc_test = roc_auc_score(y_test, lgbm_model_tuned.predict(X_test))
    print("Roc Auc Test: ", "%.3f" % lgbm_roc_auc_test)
    fpr, tpr, thresholds = roc_curve(y_test, lgbm_model_tuned.predict_proba(X_test)[:,1])
    plt.figure(figsize=(5,5))
    plt.plot(fpr, tpr, label='AUC (area = %.3f)' % lgbm_roc_auc_test )
    plt.plot([0, 1], [0, 1], 'r--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Ratio')
    plt.ylabel('True Positive Ratio')
    plt.title('ROC')
    plt.show()

  if plot_importance:
    feature_imp = pd.DataFrame({'Value': lgbm_model_tuned.feature_importances_, 'Feature': X.columns})
    plt.figure(figsize=(8, 8))
    sns.set(font_scale=1)
    sns.barplot(x="Value", y="Feature", data=feature_imp.sort_values(by="Value",
                                                                     ascending=False)[0:len(X)])
    plt.title('Features')
    plt.tight_layout()
    plt.savefig('importances.png')
    plt.show()

  if save_model:
    joblib.dump(lgbm_model_tuned, "lgbm_model_tuned.pkl")

  return lgbm_model_tuned
```
